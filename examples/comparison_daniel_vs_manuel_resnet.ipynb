```python
import pandas as pd
import matplotlib.pyplot as plt


res_net_eval_manuel = pd.read_csv('examples/resnet_eval_manuel.csv')
res_net_eval_daniel = pd.read_csv('examples/resnet_eval_daniel.csv')
res_net_eval_daniel_conv = pd.read_csv('examples/resnet_eval_daniel_conv.csv')
```

```python
plt.figure(figsize=(20, 3))

plt.plot(res_net_eval_manuel['loss'])
plt.plot(res_net_eval_daniel['loss'])
plt.plot(res_net_eval_daniel_conv['loss'])
plt.legend(['loss manuel', 'loss daniel', 'loss daniel conv'])
plt.xlabel('epochs')
plt.ylabel('loss a.u.')
plt.title('comparison loss daniel vs manuel resnet')
plt.show()
```

```python
plt.figure(figsize=(20, 3))

plt.plot(res_net_eval_manuel['kinetic_energy_loss'])
plt.plot(res_net_eval_daniel['T_loss'])
plt.plot(res_net_eval_daniel_conv['T_loss'])

plt.legend(['kinetic_energy_loss manuel', 'kinetic_energy_loss daniel', 'kinetic_energy_loss daniel conv'])
plt.xlabel('epochs')
plt.ylabel('loss a.u.')
plt.title('comparison kinetic_energy_loss daniel vs manuel resnet')
plt.show()
```

```python
plt.figure(figsize=(20, 3))

plt.plot(res_net_eval_manuel['kinetic_energy_density_loss'])
plt.plot(res_net_eval_daniel['dT_dn_loss'])
plt.plot(res_net_eval_daniel_conv['dT_dn_loss'])

plt.legend(['kinetic_energy_density_loss manuel', 'kinetic_energy_density_loss daniel', 'kinetic_energy_density_loss daniel conv'])
plt.xlabel('epochs')
plt.ylabel('loss a.u.')
plt.title('comparison kinetic_energy_density_loss daniel vs manuel resnet')
plt.show()
```
